```python
import math

#Exercise 1
import turtle
bob = turtle.Turtle()


def square(t):
    for i in range(4):
        t.fd(100)
        t.lt(90)
        print(t)
        
square(bob)

#Exercise 2
def squarel(t,length):
    """t is a turtle and length is the edge side of the square"""
    for i in range(4):
        t.fd(length)
        t.lt(90)
        print(t)

squarel(bob,36)

#Exercise 3
def polygon(t,length,n):
    for i in range(n):
        t.fd(length)
        t.lt(360/n)
        print(t)
        
polygon(bob,40,5)

#Exercise 4

def circle(t,r): 
    """this is a circle approximated by 200 steps"""
    circumf = 2*math.pi*r
    length = circumf/200
    polygon(t,length,200)
    
circle(bob,50)

#The author gives a better circle definition:
def better_circle(t,r): 
    """ This is a circle with appropriate number of steps"""
    circumf = 2*math.pi*r
    n = int(circumf/3)+3
    length = circumf/n
    polygon(t,length,n)

#Exercise 5

def arc(t,r,angle): 
    """"this is an arc using appropriate number of steps for THE WHOLE circle. 
    The angle should be considered as in degrees units
    t is the turtle, r is the radius"""
    circumf = 2*math.pi*r
    arc_length = (angle/360)*circumf
    n = int(circumf/3)+3
    steps = int((angle/360)*n)
    length = arc_length/steps    
    for i in range(steps):
        t.fd(length)
        t.lt(360/n)
        print(t)

# This is a bit different from the author's
        
arc(bob,75,270)
arc(bob,56.25,180)
arc(bob,42.18,180)
arc(bob,31.64,180)
arc(bob,23.73,180)
squarel(bob,25)
# Since the end and startpoints don't coincide,this will look like a spiral :D
#This is why we did our arc as an arc that starts FROM the last point visited.

turtle.mainloop() #this waits for the user to do something. Please, close the turtle window!
```
